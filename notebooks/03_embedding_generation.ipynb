# Embedding Generation
Generate and store embeddings for known speakers.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.inference.identify_speaker import SpeakerIdentifier
```

```python
DATA_PATH = "../data/raw"
identifier = SpeakerIdentifier(model_path="../models/speaker_encoder.pt", embeddings_path="../embeddings/speakers.json")

# Enroll all speakers found in raw data
if os.path.exists(DATA_PATH):
    for speaker in os.listdir(DATA_PATH):
        s_dir = os.path.join(DATA_PATH, speaker)
        if os.path.isdir(s_dir):
            files = [f for f in os.listdir(s_dir) if f.endswith('.wav')]
            if files:
                # Use first file for enrollment (or loop through them and average?)
                # For simplicity, using first file
                path = os.path.join(s_dir, files[0])
                identifier.enroll_speaker(speaker, path)
else:
    print("Data path not found")
```
